## Let us do some feature engineering
- creating derived features if possible(from sc_height,sc_width new feature such as total resolution ...same with pixels)
- removing feature with less correlation with "price_range" (fc,clock_speed)
- shuffling and separating the features,labels

```python
##let us remove some features
df1 = df.drop(columns=['fc','clock_speed','m_dep'])
```

```python
len(df.columns),len(df1.columns)
```

```python
df1.head()
```

```python
##shuffling out the dattafram

df1 = df1.sample(frac = 1)
df1.head()
```

```python
##separating features,labels
features=df1.drop("price_range",axis=1)
labels=df1['price_range'].copy()
labels.head()
features.head()
```

```python
##checking by our custom code for na values
ft_na = [ft for ft in features.columns if features[ft].isna().sum()>0]
ft_na
```
